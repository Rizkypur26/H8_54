```python
from wordcloud import WordCloud, STOPWORDS
```

```python
import pandas as pd
import numpy as np
```

```python
with open ("alice_novel.txt")as f:
    alice_novel = f.read()
```

```python
alice_novel[:200]
```

```python
alice_wc = WordCloud(background_color = 'black', max_words = 2000, stopwords=STOPWORDS)
```

```python
alice_wc.generate(alice_novel)
```

```python
import matplotlib.pyplot as plt

plt.imshow(alice_wc)
plt.show()
```

```python
map = folium.Map([-6.229728, 106.8132282], zoom_start = 14)

lat = -6.227366
long = 106.7949765

markers = folium.map.FeatureGroup()
markers.add_child(folium.CircleMarker([lat,long], radius = 30, color = 'green', fill=True, fill_color = 'yellow'))
map.add_child(markers)
```
